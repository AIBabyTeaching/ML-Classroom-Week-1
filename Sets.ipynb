# Handouts 6: Sets

A set is an unordered collection of items. Every set element is unique (no duplicates) and must be immutable (cannot be changed).

However, a set itself is mutable. We can add or remove items from it.

Sets can also be used to perform mathematical set operations like union, intersection, symmetric difference, etc.

## 1. Creating Sets

```python
# Different types of sets in Python

# set of integers
my_set = {1, 2, 3}
print(my_set)

# set of mixed datatypes
my_set = {1.0, "Hello", (1, 2, 3)}
print(my_set)
```

```python
# set cannot have duplicates
my_set = {1, 2, 3, 4, 3, 2}
print(my_set)

# we can make set from a list
my_set = set([1, 2, 3, 2])
print(my_set)

# set cannot have mutable items
# here [3, 4] is a mutable list
# this will cause an error.
my_set = {1, 2, [3, 4]}
```

```python
# Distinguish set and dictionary while creating empty set

# initialize a with {}
a = {}
# a = {1, 2, 3, 4}

# check data type of a
print(type(a))

# initialize a with set()
a = set()

# check data type of a
print(type(a))
```

## 2. Modifying a set

```python
# initialize my_set
my_set = {1, 3}
print(my_set)

# my_set[0]
# if you uncomment the above line
# you will get an error
# TypeError: 'set' object does not support indexing

# add an element
my_set.add(2)
print(my_set)

# add multiple elements
my_set.update([2, 3, 4, 5])
print(my_set)

# add list and set
my_set.update([4, 5], {1, 6, 8}, (9, 10))
print(my_set)
```

## 3. Removing elements from a set

```python
# Difference between discard() and remove()

# initialize my_set
my_set = {1, 3, 4, 5, 6}
print(my_set)

# discard an element
# Output: {1, 3, 5, 6}
my_set.discard(4)
print(my_set)

# remove an element
# Output: {1, 3, 5}
my_set.remove(6)
print(my_set)

# discard an element
# not present in my_set
# Output: {1, 3, 5}
my_set.discard(2)
print(my_set)

# remove an element
# not present in my_set
# you will get an error.
# Output: KeyError

my_set.remove(2)
```

```python
# initialize my_set
# Output: set of unique elements
my_set = set("HelloWorld")
print(my_set)

# pop an element
# Output: random element
print(my_set.pop())

# pop another element
my_set.pop()
print(my_set)

# clear my_set
# Output: set()
my_set.clear()
print(my_set)
```

## 4. Union

```python
# Set union method
# initialize A and B
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

# use | operator
print(A | B)
print(A.union(B))
print(B.union(A))
```

## 5. Intersection

```python
# Intersection of sets
# initialize A and B
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

# use & operator
print(A & B)
print(A.intersection(B))
print(B.intersection(A))
```

### 6. Difference

```python
# Difference of two sets
# initialize A and B
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

# use - operator on A
print(A - B)
print(A.difference(B))
print(B.difference(A))
```

### 7. Symmetric Difference

Symmetric Difference of A and B is a set of elements in A and B but not in both (excluding the intersection).

```python
# Symmetric difference of two sets
# initialize A and B
A = {1, 2, 3, 4, 5}
B = {4, 5, 6, 7, 8}

# use ^ operator
print(A ^ B)
print(A.symmetric_difference(B))
print(B.symmetric_difference(A))
```

### 8. Subset

```python
A = {1,2,3,10}
B = {1,2,3,4,5,6}
print(A.issubset(B))
print(B.issubset(A))
```

### 9. Set Membership Test

```python
my_set = set("apple")

# check if 'a' is present
print('a' in my_set)

# check if 'p' is present
print('p' not in my_set)
```

### 10. Iterating Through a Set

```python
numbers = {5,6,7,8,9,10}
for number in numbers:
    print(number)
```

### 11.  Len( ), min( ), max( ), sum( ), sorted( )

```python
numbers = {40, 3, 10, 50, 7, 90, 83}
print("Orginal Set:", numbers)
print("length of set is", len(numbers))
print("Min of set is", min(numbers))
print("Max of set is", max(numbers))
print("Sum of list is", sum(numbers))
print("The sorted list: ", sorted(numbers))
```
