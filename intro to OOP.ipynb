## 3. Constructors (are  used for instantiating an object)

```python
class Employee:
    # constructor
    def __init__(self, name, working_hrs, hour_rate=90):
        # Assign to self object (instance attributes)
        self.name = name
        self.working_hrs = working_hrs
        self.hour_rate = hour_rate

    # The method should have at least one parameter which is the object it self
    def calc_Salary(self):
        salary = self.working_hrs * self.hour_rate
        print("The Salary is:", salary)
        return salary

# obj 1
employee_1 = Employee("Ahmed Essam", 35)
employee_1.calc_Salary()

# obj 2
employee_2 = Employee('Sarah Ashraf', 40, 120)
employee_2.calc_Salary()

# obj 2
# hrs = 90
# rate = 110
# employee_2 = Employee("Fareda Ayman", hrs, rate)
# print(employee_2.name)
# print(employee_2.working_hrs)
# print(employee_2.hour_rate)
# employee_2.calc_Salary()
```

## 4. Data Validation (using assert statement)

```python
class Employee:
    # constructor with data type for parameters
    def __init__(self, name, working_hrs, hour_rate):

        # make sure that hour rate is always +ve number
        assert hour_rate >= 0

        # Assign to self object (instance attributes)
        self.name = name
        self.working_hrs = working_hrs
        self.hour_rate = hour_rate

    # The method should have at least one parameter which is the object it self
    def calc_Salary(self):
        salary = self.working_hrs * self.hour_rate
        print("The Salary is:", salary)
        return salary

# obj 1
employee_1 = Employee("Ahmed Essam", "35", -1)
print(employee_1.name)
print(employee_1.working_hrs)
print(employee_1.hour_rate)

# obj 2
hrs = 90
rate = 200
employee_2 = Employee("Fareda Ayman", hrs, rate)
print(employee_2.name)
print(employee_2.working_hrs)
print(employee_2.hour_rate)
employee_2.calc_Salary()
```

## 8. Class methods VS. Static methods

```python
from datetime import date as dt
class Employee:
   def __init__(self, name, age):
      self.name = name
      self.age = age

   @staticmethod
   def isAdult(age):
      if age > 18:
         return True
      else:
         return False

   @classmethod
   def emp_from_year(emp_class, name, year):
      return emp_class(name, dt.today().year - year)

   def __str__(self):
      return 'Employee Name: {} and Age: {}'.format(self.name, self.age)

e1 = Employee('Dhiman', 25)
print(e1)

e2 = Employee.emp_from_year('Subhas', 1987)
print(e2)

print(Employee.isAdult(25))
print(Employee.isAdult(16))
```

## 9. private attributes

For the purpose of data encapsulation, most object oriented languages use getters and setters method. This is because we want to hide the attributes of a object class from other classes so that no accidental modification of the data happens by methods in other classes.

As the name suggests, getters are the methods which help access the private attributes or get the value of the private attributes and setters are the methods which help change or set the value of private attributes.

```python
# Making Getters and Setter methods
class Celsius:
    def __init__(self, temperature=0):
        self.set_temperature(temperature)

    def to_fahrenheit(self):
        return (self.get_temperature() * 1.8) + 32

    # getter method
    def get_temperature(self):
        # An underscore _ at the beginning is used to denote private variables in Python.
        return self._temperature

    # setter method
    def set_temperature(self, value):
        if value < -273.15:
            raise ValueError("Temperature below -273.15 is not possible.")
        self._temperature = value


# Create a new object, set_temperature() internally called by __init__
human = Celsius(37)

# Get the temperature attribute via a getter
print(human.get_temperature())

# Get the to_fahrenheit method, get_temperature() called by the method itself
print(human.to_fahrenheit())

# new constraint implementation
human.set_temperature(-300)

# Get the to_fahreheit method
print(human.to_fahrenheit())
```
