# Handouts 5: Tuples

A tuple is a collection of objects which ordered and immutable. Tuples are sequences, just like lists. The differences between tuples and lists are, the tuples cannot be changed unlike lists and tuples use parentheses, whereas lists use square brackets.

### 1. Creating Tuple


```python
# Different types of tuples

# Empty tuple
my_tuple = ()
print(my_tuple)

# Tuple having integers
my_tuple = (1, 2, 3)
print(my_tuple)

# tuple with mixed datatypes
my_tuple = (1, "Hello", 3.4)
print(my_tuple)

# nested tuple
my_tuple = ("mouse", [8, 4, 6], (1, 2, 3))
print(my_tuple)
```

```python
# A tuple can also be created without using parentheses. This is known as tuple packing.
my_tuple = (3, 4.6, "dog", "Cat")
print(my_tuple)

# tuple unpacking is also possible
a, b, c, d = my_tuple

print(a)      # 3
print(b)      # 4.6
print(c)      # dog
print(d)
```

```python
# Creating a tuple with one element is a bit tricky.
# Having one element within parentheses is not enough.
# We will need a trailing comma to indicate that it is, in fact, a tuple.

my_tuple = ("hello")
print(type(my_tuple))  # <class 'str'>

# Creating a tuple having one element
my_tuple = ("hello",)
print(type(my_tuple))  # <class 'tuple'>

# Parentheses is optional
my_tuple = "hello",
print(type(my_tuple))  # <class 'tuple'>
```

### 2. Access Tuple Elements

#### 2.1 Indexing

```python
# Accessing tuple elements using indexing
my_tuple = ('p','e','r','m','i','t')

print(my_tuple[0])
print(my_tuple[5])

# IndexError: list index out of range
# print(my_tuple[6])

# Index must be an integer
# TypeError: list indices must be integers, not float
# my_tuple[2.0]

# nested tuple
n_tuple = ("mouse", [8, 4, 6], (1, 2, 3))

# nested index
print(n_tuple[2][1])
print(n_tuple[1][1])
```

#### 2.2 Negative indexing

```python
# Negative indexing for accessing tuple elements
my_tuple = ('p', 'e', 'r', 'm', 'i', 't')

print(my_tuple[-1])
print(my_tuple[-6])
```

#### 2.3 Slicing

```python
# Accessing tuple elements using slicing
my_tuple = ('p','r','o','g','r','a','m','i','z')

print(my_tuple[1:4])

print(my_tuple[:-7])

print(my_tuple[7:])

print(my_tuple[:])
```

### 3. Changing a Tuple

Unlike lists, tuples are immutable.

This means that elements of a tuple cannot be changed once they have been assigned. But, if the element is itself a mutable data type like a list, its nested items can be changed.

We can also assign a tuple to different values (reassignment).

```python
# Changing tuple values
my_tuple = (4, 2, 3, [6, 5])

# TypeError: 'tuple' object does not support item assignment
# my_tuple[1] = 9

# However, item of mutable element can be changed
my_tuple[3][0] = 9    # Output: (4, 2, 3, [9, 5])
print(my_tuple)

# Tuples can be reassigned
my_tuple = (4, 2, 3, 7, [6, 5])

print(my_tuple)
```

## 4. Tuple Operators

```python
# Concatenation
print((1, 2, 3) + (4, 5, 6))

# Repeat
print(("Good",) * 10)
```

### 5. Deleting a Tuple

```python
# Deleting tuples
my_tuple = ('p', 'r', 'o', 'g', 'r', 'a', 'm', 'i', 'z')

# can't delete items
# TypeError: 'tuple' object doesn't support item deletion
# del my_tuple[3]

# Can delete an entire tuple
del my_tuple

# NameError: name 'my_tuple' is not defined
print(my_tuple)
```

### 6. Tuple methods

```python
my_tuple = ('a', 'p', 'p', 'p', 'l', 'e',)

print(my_tuple.count('p'))
print(my_tuple.index('l'))
```

### 7. Tuple Membership Test

```python
# Membership test in tuple
my_tuple = ('a', 'p', 'p', 'l', 'e',)

# In operation
print('a' in my_tuple)
print('b' in my_tuple)

# Not in operation
print('g' not in my_tuple)
```

### 8. Iterating Through a Tuple

```python
# Using a for loop to iterate through a tuple
names = ('John', 'Kate', 'Mary', 'Andrew')
for name in names:
    print(name)
```

### 10. Tuple functions

#### 10.1 len( )

```python
names = ('John', 'Kate', 'Mary', 'Andrew')
len(names)
```

#### 10.2 max(), min()

```python
numbers = (100, 10, 3, 60, 150)
print ("The max of tuple is:",  max(numbers))
print ("The min of tuple is:", min(numbers))
```

#### 10.3 Convert list into tuple using tuple( )

```python
numbers_list = [5,6,7,8,9,10]
numbers_tuple = tuple(numbers_list)
print(type(numbers_tuple))
```

#### 10.4 Convert tuple into list using list( )

```python
numbers_tuple = (5,6,7,8,9,10)
numbers_list = list(numbers_list)
print(numbers_list)
```
